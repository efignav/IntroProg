## BUCLES

Un bucle es una estructura de control que repite una serie de instrucciones. Hay dos tipos:

- while
- for

### Bucle ***while***

Repite una serie de instrucciones siempre que se cumpla una condicion logica.

**Ejemplo 1**: imprimir los numeros seguidos desde 0 hasta 4.

```python
print(0)
print(1)
print(2)
print(3)
print(4)
```

```python
contador = 0
while (contador < 5):
    print(contador)
    contador += 1
```

La variable contador se incremente en 1 en cada repeticion hasta que alcanza un valor específico.

**Ejemplo 2**: Suma de números pares:

Este bucle *while* suma números pares hasta que el numero no supera un valor dado.

```python
suma = 0
numero = 2
limite = 10

while numero < limite:
    suma += numero
    numero += 2

print(suma)
```

**Ejemplo 3**: Suma de varios numero ingresados porr el usuario.

```python
i = 1
total = 0
while (i <= 3):
  numero = int(input("Ingrese un numero entero: "))
  total += numero
  i += 1
  print("\tEl total es ", total)
```

```python
# Factorial de un número:
# Este bucle while calcula el factorial de un número dado.

numero = int(input("Ingresa un número entero: "))
factorial = 1
i = 1

while i <= numero:
    factorial *= i
    i += 1

print(f"El factorial de {numero} es {factorial}")
```

```python
# Lista de números pares:
# Este bucle while crea una lista de números pares en un rango dado.

inicio = 2
final = 20
numeros_pares = []

while inicio <= final:
    numeros_pares.append(inicio)
    inicio += 2

print(numeros_pares)
```

Ejemplo 4: En el sgte ejemplo se compara el valor de la variable respuesta. Mientras la variable respuesta sea igual a "si" el bucle seguir ejecutandose.

```python
dic = {}
respuesta = input("Desea ingresar informacion (si/no)? ")
while respuesta == "si" or respuesta == "SI" or respuesta == "Si" or respuesta == "sI":
  codigo = input("Codigo del producto:")
  nombre = input("Nombre del producto:")
  precio = float(input("Precio del producto:"))

  dic[codigo] = [nombre, precio]
  respuesta = input("Desea agregar mas(si/no)? ")

print(dic)
```

Bucle ***while/else***

La parte *else* se ejecutara cuando no se cumpla la condicion logica.

```python
dic = {}
respuesta = input("Desea ingresar informacion (si/no)? ")
while respuesta == "si":
  codigo = input("Codigo del producto:")
  nombre = input("Nombre del producto:")
  precio = float(input("Precio del producto:"))

  dic[codigo] = [nombre, precio]
  respuesta = input("Desea agregar mas(si/no)? ")
else:
  print("Registro completo!")

print(dic)

```

Sentencia ***break***

Interrumpe la ejecucion del bucle.

```python
numero = 0
while numero <= 10:
    print("Valor actual de numero:", numero)
    numero += 1
    if numero == 6:
        break
```

```python
# Adivina el número:
# En este juego, el usuario intenta adivinar un número generado aleatoriamente.

import random

numero_secreto = random.randint(1, 10)
intentos = 0

while True:
    adivina = int(input("Adivina el número secreto que esta entre 1 y 10 : "))
    intentos += 1

    if adivina == numero_secreto:
        print(f"¡Correcto! Adivinaste en {intentos} intentos.")
        break
    elif adivina < numero_secreto:
        print("El número es mayor que", adivina)
    else:
        print("El número es menor que", adivina)
```

```python
# Bucle While para Encontrar un Número Específico:

numero_buscado = 7
contador = 1

while contador <= 10:
    if contador == numero_buscado:
        print(f"¡Encontré el número {numero_buscado}!")
        break
    contador += 1
```

```python
# Bucle While para Encontrar Números Pares:

numero = 1

while True:
    if numero % 2 == 0:
        print(f"Encontré un número par: {numero}")
    numero += 1
    if numero > 20:
        break
```

```python
# Bucle While con Entrada de Usuario:

while True:
    entrada = input("Ingresa 'salir' para terminar: ")
    if entrada.lower() == 'salir':
        break
```

```python
# Bucle While para Sumar Números hasta un Límite:

limite = 100
suma = 0
contador = 1

while True:
    suma += contador
    contador += 1
    if suma > limite:
        break

print(f"La suma de números hasta {limite} es: {suma}")
```

Sentencia ***continue***

Mueve el control de la ejecucion al inicio de la presente iteracion.



```python
numero = 0
while numero <= 10:
    numero += 1
    if numero % 2 == 0:
        continue

    print("Valor actual de numero:", numero)
```

```python
# Bucle While con Uso de continue:

contador = 1

while contador <= 10:
    if contador % 2 != 0:
        contador += 1
        continue  # Salta los números impares
    print(f"Número par: {contador}")
    contador += 1
```

### Bucle ***for***

Recorre los elementos de un objeto ***iterable*** como una tupla, lista, conjunto o diccionario.

```python
numeros = [3, 8, 9, 1, 4, 10]

for n in numeros:
  print(n)
```

```python
mitup = (80, 90, 10)

for n in mitup:
  print(n)
```

```python
conjunto = {30, 18, 140}

for z in conjunto:
  print(z)
```

Tarea: Usar ***for*** con un diccionario.

```python
dicion = {1:'A', 2:'B', 3:'C'}

for x, y in dicion.items():
  print(">>", x, "=",y)
```

```python
dias = ["Domingo", "Lunes", "Martes", "Miercoles"]
for d in dias:
    print("Dia:", d)
```

### ***for - range***

```python
for n in range(4):
    print(n)
```

```python
for num in range(15, 15+7):
    print("Los dias son:", num)
```

```python
for num in range(15, 22, 3):
    print("Los dias son:", num)
```
