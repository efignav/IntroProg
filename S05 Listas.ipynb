METODO pop()

Elimina un elemento, por medio del indice de una lista. El problema que se presenta es que si hay muchos elementos, tenemos que averiguar cual es su indice.

```python
listasa = [x for x in range(11, 121, 2)]
```

```python
print(listasa)
print(len(listasa))
```

Usando el metodo pop  eliminar el numero 47

```python
listasa.index(47)
```

```python
listasa.pop( listasa.index(47) )
```

```python
print(listasa)
```
