```python
# diccionario vacio
dic = {}
print('diccionario vacio :', dic)
```

```python
# ejemplo 1
dic1 = {'nombre': "juan" , 'edad': 20 , 'telf': 65943258}
print(dic1)
```

```python
# ejemplo 2
dic2 = {'nombre':"juan" , 'edad':20, 'cursos': ['matematica', 'lenguaje'] }
print(dic2)
```

```python
# acceso
print('Nombre:', dic2['nombre'])
print('Edad:', dic2['edad'])
print('Cursos :',  dic2['cursos'])
print('Lenuaje :',  dic2['cursos'][1])
```

```python
# tamano del diccionario
print('longitud dic1:', len(dic1) ) # length: longitud
print('longitud dic2:', len(dic2) ) 
```

```python
# claves
clave = dic1.keys()
print("claves de dic1:", clave)
print()
clave = dic2.keys()
print("claves de dic2:", clave)
```

```python
# # valores
valor = dic1.values()
print("los valores de dic1 son:", valor)
print()
valor = dic2.values()
print("los valores de dic2 son:", valor)
```

```python
# modificacion
dic4 = {'nombre':"ivan", 'edad':20}
print(dic4)
print()
dic4['edad'] = 17
print(dic4)
```

# Metodos de un diccionario

```python
# agregar
dic4 = {'nombre':"ivan", 'edad':20}
print(dic4)
print()
dic4['apellido'] = 'gonzales'
print(dic4)
dic4['sexo'] = 'F'
print(dic4)
dic4['telefono'] = '89675432'
print(dic4)
```

```python
# eliminar
dic5 = {'nombre':"jose", 'edad': 17}
print(dic5)
print()

dic5.pop('edad')
print(dic5)
print()

dic5.pop('nombre')
print(dic5)
```

```python
# duplicar un diccionario
dic6 = {'producto':'azucar', 'precio': 3.5}
print("dic6", dic6)
print()

dic6_dup = dic6.copy()
print('dic6_dup :', dic6_dup)

dic6_dup.clear()
print('dic6_dup :', dic6_dup)
```

## Acceso

```python
dic2 = {'nombre':"juan", 'edad':20, 'cursos':['mate', 'lenguaje']}
print()
print("dic2", dic2)

print()
print("el nombre es:", dic2['nombre'])
print("su edad es:", dic2['edad'])
print("y esta llevando los cursos:", dic2['cursos'])
```

```python
dic2 = {'nombre':"juan", 'edad':20, 'cursos':['mate', 'lenguaje']}
print(dic2)
# print(dic2['apellido'])
print(dic2.get('apellido'))

print(dic2.get('apellido', 'Clave apellido no existe' ))
```

```python
# update

dic8 = {'prod':'atun', 'prec': 4.5, 'marca':'fanny'}
print()
print(dic8)

dic8.update({'prec': 5.2, 'marca': 'rocio'})
print()
print(dic8)
```
