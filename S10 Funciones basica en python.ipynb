## Parametros de un función

Si una función se invoca sin especificar algún parámetro esta puede trabajar con uno *por default.*

Por ejemplo, la función **print** tiene un parámetro *end* para especificar el carácter de fin de linea, que por default es un salto de linea.

```python
print("HOLA")
print("Buenos dias ", "mundo")
```

Podemos especificar que tenga un caracter de tabulado del siguiente modo:

```python
print("HOLA", end= " ")
print("Buenos dias ", "mundo")
```

```python
print("Buenos dias", "mundo", sep=' = ')
```

```python
print("Buenos dias", "mundo", sep=' + ', end="\n\t")

print("Hoy es miercoles 1ro", "de noviembre", sep=' : ')
```

# len()
Para averiguar el tamano de una coleccion, por ejemplo, una lista.

```python
lista1 = ['a', 'b', 'c', 'x']

longitud_lista1 = len(lista1)
print ("Longitud de la lista1 :", longitud_lista1 )

nombre = 'T & C Week'
print ("Longitud de \'T & C Week\'  :", len(nombre) )
```

```python
tup1 = ('a', 'b', 'c', 'xx', 13)

longitud_tup1 = len(tup1)
print ("Longitud de tup1 :", longitud_tup1 )
```

```python
c1 = {'a', 'b', 'c', 'a', 'a'}

print ("Longitud de c1 :", len(c1) )
```

# min()

Halla el valor minimo de una coleccion.

```python
lista2 = [18, 12, 0, 20, 19]

print("La nota minima es", min(lista2) )
```

Se desea obtener el promedio de las 4 notas mas altas de lista2.

```python
lista2 = [18, 12, 0, 20, 19]
nota_baja = min(lista2)
print("Nota mas baja:", nota_baja)

suma_notas = sum(lista2)
print("Suma de todas las notas:", suma_notas)
suma_menos_nota_baja = suma_notas - nota_baja

print("Suma de todas las notas MENOS la nota mas baja:", suma_menos_nota_baja)
promedio = suma_menos_nota_baja / (len(lista2) - 1)
promedio_original = suma_menos_nota_baja / (len(lista2))
print("Promedio CON la nota mas baja", round(promedio_original,0))
print("Promedio sin la nota mas baja", round(promedio, 0))
```

```python
tupla2 = (18, 12, 0, 20, 19)
nota_baja = min(tupla2)
print("Nota mas baja:", nota_baja)

suma_notas = sum(tupla2)
print("Suma de todas las notas:", suma_notas)
suma_menos_nota_baja1 = suma_notas - nota_baja
print(tupla2.index(nota_baja))
lista2 = list(tupla2)
print(lista2.index(min(lista2)))
lista2.pop(lista2.index(min(lista2)))
print(lista2)
```
