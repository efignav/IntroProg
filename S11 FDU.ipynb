Una funcion que promedia en forma mas flexible.

```python
def nota_periodo(calificaciones):
    suma = sum(calificaciones)
    prom = suma / len(calificaciones)

    return round(prom)
```

```python
nota_periodo({20, 10, 20, 20})
```

```python
alumnos = {'alumno1': [16, 15], 'alumno2': [12, 18], 'alumno3': [20, 0]}

for alumno, notas in alumnos.items():
    print(alumno, notas)
```

```python
alumnos_a = {'alumno1': [10, 10, 10], 'alumno2':[12, 20], 'alumno3':[20, 20, 18]}

for alumno, notas in alumnos_a.items():
    print("Promedio del",alumno, "=",round(nota_periodo(notas), 1 ) )
```

```python
alumnos_b = {'alumno1': [12, 20, 13], 'alumno2':[10, 18, 16], 'alumno3':[19, 12, 18]}

for alumno, notas in alumnos_b.items():
  # print(alumno, sum(notas)/2)
  # print(alumno, sum(notas)/len(notas))
  # print(alumno, round(sum(notas)/len(notas)))
  print(alumno, nota_periodo(notas ) )
```

Si se quiere saber que calificaciones deben obtener para aprobar un curso, sabiendo las 2 primeras.

```python
def cuanto_me_falta(lista_notas):
    f = 50 - sum(lista_notas)
    return f / (4-len(lista_notas))
```

```python
misnotas = [18, 17]
x = cuanto_me_falta(misnotas)
print("Me falta", x, "por cada periodo")
```
