Funcion para abrir archivo de texto

```python
def funcion_x(file, modo):
    if modo == "r":
      print("Asegurese que ", file, " existe")
      return

    miarchivo = open(file, modo)
    # miarchivo.close()
    print(miarchivo.name)
    print(miarchivo.closed)
    print(miarchivo.mode)
```
