```python
n = "Enrique"
e = 35
a = 1.7

print(f"Tu nombre es {n} y tu edad es {e} y mides {a}" )
```

```python
n1="Pedro"
n2="Anaximenes"
n3="Eva"

print("{0:>10}".format(n1))
print("{0:>10}".format(n2))
print("{0:>10}".format(n3))
```
