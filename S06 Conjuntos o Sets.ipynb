# OPERACIONES SOBRE CONJUNTOS
- union
- intersection

# **union**

```python
A = {1, 2, 3, 4}
B = {5, 4, 2, 7, 8}

print(A, B)
```

```python
C = A | B
print(C)
```

```python
C2 = A.union(B)

print(C2)
```

# **intersection**

```python
A = {1, 2, 3, 4}
B = {5, 4, 2, 7, 8}

print(A, B)
```

```python
D = A & B
print(D)
```

```python
D2 = A.intersection(B)
print(D2)
```

# **difference**

```python
A = {1, 2, 3, 4}
B = {5, 4, 2, 7, 8}

print(A, B)
```

```python
E = A - B
print(E)
```

```python
E2 = A.difference(B)
print(E2)
```

# **symmetric_difference**

```python
A = {1,2,3,4}
B = {5,4,2,7,8}

print(A, B)
```

```python
F = A ^ B
print(F)
```

```python
F2 = A.symmetric_difference(B)

print(F2)
```

La diferencia simetrica es equivalente a:

```python
A.union(B) - A.intersection(B)
```

# OPERACIONES SOBRE CONJUNTOS

- issubset()
- issuperset
- isdisjoint

***issubset()***

```python
A = {1,2,3,4,5,6,7,8,9}

B = {5,4,2,7,8}

A <= B
```

```python
A.issubset(B)
```

***issuperset()***

```python
A = {1,2,3,4,5,6,7,8,9}
B = {5,4,2,7,8}

A >= B
```

```python
A.issuperset(B)
```

***isdisjoint()***

```python
A = {1,2,3,4,5,6,7,8,9}
B = {5,4,2,7,8}

A.isdisjoint(B)
```
