# Operadores logicos

# and

```python
a = 8
b = 3
c = 12
d = 6

print(a == b)  # False
print(c > d)   # True
print()

(a == b) and (c > d)
```

```python
a = 8
b = 3
c = 12
d = 6

print(a != b) # V
print(c > d)  # V
print()

(a != b) and (c > d)

# si es True hacer una cosa
# si es False hacer otra cosa
```

# or

```python
a = 8
b = 3
c = 12
d = 6

print(a == b)   # F
print(c > d)    # V
print()

(a == b) or (c > d)
```
